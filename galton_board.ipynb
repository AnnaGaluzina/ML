```python
import seaborn as sns  # библиотека для визуализации
import random


def galton_board(balls, levels):
    results = {}

    for ball in range(balls):
        position = 0
        for level in range(levels):
            position += random.choice([-1, 1])
        position /= levels
        position = round(position, 1)
        if position in results:
            results[position] += 1
        else:
            results[position] = 1

    return results

balls = 10000; levels = 10
result = galton_board(balls, levels)
print(result)
# visualization code
sns.barplot(x=list(result.keys()), y=list(result.values()))
```
